```python
weapon = [[454,0,0,0,0,36,221,"满精匣里"],  
          [454,0,0,0,0,32,221,"4精匣里"],  
          [608,0,66.2,0,20,0,0,"1精专武"],
          [454,37,0,36,0,0,0,"满精角斗对单"],
          [565,21,0,60.6,0,0,0,"满精千岩3人"],
#           [565,28,0,27.6+44,0,0,0,"满精千岩4人"],
          [674,22,0,22.4,0,12,0,"和濮鸢"],
          ]

Yiwu_set = [[4780,311,168,0.466,62.2],
            [4780,311,0.466,0.466,62.2]]

num_citiao = 30
```

```python
def get_damages():
#     套装防岩爆
    
    damage_result = [] 
    big_damage_result = []

    for n, aWeapon in enumerate(weapon):
        best_damage = 0
        best_equip = [0, 0, 0, 0, 0, 0]
        mianban = []
        big_damage = 0
        big_damage_mianban = []
        big_damage_equip = []
        weaponResult = []
        weaponBigRe = []
        for cri_num in range(num_citiao):
            for criD_num in range(num_citiao - cri_num):
                for str_num in range(num_citiao - cri_num - criD_num):
                    for mastery_num in range(num_citiao - cri_num - criD_num - str_num):
                        for aSet in range(1,3):
                            hp_num = num_citiao - cri_num - criD_num - str_num - mastery_num

                            if aSet == 1:
                                mastery = aWeapon[6] + 18*mastery_num+187
                                hp = basic_hp*(1+(hp_num*5 +aWeapon[4])/100) + 4780 + 1000
                                cri = basic_cri+cri_num*3.3+aWeapon[1]
                                if cri > 100:
                                    cri = 100
                                criD = basic_criD + criD_num*7 +aWeapon[2]+62.2

                                if "1精专武" in aWeapon[7]:

                                    Str = (basic_str+aWeapon[0])*((str_num*5.5+aWeapon[3])/100+1)+311 + hp*(0.018+0.0626)
                                else:
                                    Str = (basic_str+aWeapon[0])*((str_num*5.5+aWeapon[3])/100+1)+311 + hp*(0.0626)

                                buff = 46.6 + 33 + aWeapon[5]+ set_fire_buff
                                
                            else:
                                mastery = aWeapon[6] + 18*mastery_num
                                hp = basic_hp*(1+(hp_num*5 +aWeapon[4]+46.6)/100) + 4780 + 1000
                                cri = basic_cri+cri_num*3.3+aWeapon[1]
                                if cri > 100:
                                    cri = 100
                                criD = basic_criD + criD_num*7 +aWeapon[2]+62.2

                                if "1精专武" in aWeapon[7]:

                                    Str = (basic_str+aWeapon[0])*((str_num*5.5+aWeapon[3])/100+1)+311 + hp*(0.018+0.0626)
                                else:
                                    Str = (basic_str+aWeapon[0])*((str_num*5.5+aWeapon[3])/100+1)+311 + hp*(0.0626)

                                buff = 46.6 + 33 + aWeapon[5]+ set_fire_buff

                            master_buff = 0.00000005*mastery*mastery*mastery + 0.00001*mastery*mastery + 0.169*mastery + 0.0281

                            damage = (Str*Skill)*(1+buff/100)*(1+cri*criD/10000)*0.5*0.9*(1.15+master_buff/100)*1.5
                            criDamage = (Str*Skill)*(1+buff/100)*(1+criD/100) *0.5*0.9*(1.15+master_buff/100)*1.5

                            if damage > best_damage and cri >50 and criD_num + cri_num  <= 20  and cri_num < 15 :
                                best_damage = damage
                                best_equip = [cri_num , criD_num , str_num , hp_num ,n ,mastery_num,aSet]
                                mianban = [Str, cri, criD, hp, mastery]
                            if criDamage > big_damage and cri >= 50 and criD_num + cri_num <= 20 :
                                big_damage = criDamage
                                big_damage_mianban = [Str, cri, criD, hp, mastery]
                                big_damage_equip = [cri_num , criD_num , str_num , hp_num ,n ,mastery_num,aSet ]
        
        weaponResult.append(best_damage)
        weaponResult.append(best_equip)
        weaponResult.append(mianban)
        
        weaponBigRe.append(big_damage)
        weaponBigRe.append(big_damage_equip)
        weaponBigRe.append(big_damage_mianban)
        
        damage_result.append(weaponResult)
        big_damage_result.append(weaponBigRe)
        

    return damage_result, big_damage_result
```

```python
def Datasort(weapon_result):
    for i in range(len(weapon_result)):
        for j in range(i+1,len(weapon_result)):
            if weapon_result[j][0] > weapon_result[i][0]:
                aRe = []
                aRe = weapon_result[i]
                weapon_result[i] = weapon_result[j]
                weapon_result[j] = aRe
                
    return weapon_result
```

```python
damage_result, big_damage_result = get_damages()
damage_result = Datasort(damage_result)
big_damage_result = Datasort(big_damage_result)

for i,aResult in enumerate(damage_result):
    if i < 6:
        print("第"+str(i+1)+"输出面板：", aResult[2])

        print("暴击词条：", aResult[1][0])
        print("爆伤词条：", aResult[1][1])
        print("攻击词条：", aResult[1][2])
        print("生命词条：", aResult[1][3])
        print("精通词条：", aResult[1][5])
        print("配装武器：", weapon[aResult[1][4]][7])
        
        if aResult[1][6] == 1:
            print("配装精通沙漏")
        else:
            print("配装生命沙漏")
        print("第"+str(i+1)+"输出期望", aResult[0])
        
        if i == 0:
            best_damage = aResult[0]
        if i > 0:
            print("与第1输出期望差距", round((best_damage/aResult[0]-1)*100,2), "%")
        print("--------------------------------------------------------------------")

print("\n")
print("\n")
```
