```python
import numpy as np

hua = [[9.7, 0, 8.7, 53, 4.5],
       [0, 15.5, 16.9, 33,0],
       [7.8,25.6,0,0,0],
       [11.3,0,0,33,11]]

mao = [[0,17.1,0,311,19.4],
       [14.8,0,5.8,311,0],
       [0,21,9.9,311,12.3]]

sha = [[3.5,13.2,46.6,0,22.7],
       [13.6,0,46.6,0,6.5],
       [7.4,35.7, 0,0,51.8],
       [16.3,5.4,4.1,0,51.8]]

tou = [[0,62.2,5.3,53,9.1],
       [31.1,10.9,0,0,10.4]]

bei = [[7.4,5.4,0,29,0,46.6],
       [3.5,0,46.6,0,18.1,0]]

# weapon = [[674,22.1,0,3.2,0,0],
#           [454,36.8,0,32,0,0],
#           [510,12,0,0,32,45.9],
#           [674,8,0,0,0,37],
#           [674,22.1,0,11,6,0],
#           ]
weapon = [[674,22.1,0,3.2,0,0],
          [454,36.8,0,32,0,0],
          [510,12,0,0,32,45.9],
          [674,8,0,0,0,37],
          ]         

names = ["不叠鸟枪","满精角斗（群）","满精鱼叉","薄荷","平均鸟枪","满精角斗（单）"]

basic_str = 337
basic_cri = 5
basic_criD = 50
basic_char = 132
```

```python
def get_damages():
    best_damage = 0
    best_equip = [0, 0, 0, 0, 0, 0]
    mianban = []
    big_damage = 0
    big_damage_mianban = []
    big_damage_equip = []
    for i,aHua in enumerate(hua):
        for j, aMao in enumerate(mao):
            for k, aSha in enumerate(sha):
                for l, atou in enumerate(tou):
                    for m, aBei in enumerate(bei):
                        for n, aWeapon in enumerate(weapon):
                            str = (basic_str+aWeapon[0])*((aBei[2]+aHua[2]+aMao[2]+aSha[2]+atou[2]+aWeapon[3])/100+1)+(aBei[3]+aHua[3]+aMao[3]+aSha[3]+atou[3])
                            cri = basic_cri+aBei[0]+aHua[0]+aMao[0]+aSha[0]+atou[0]+aWeapon[1]
                            criD = basic_criD + aBei[1]+aHua[1]+aMao[1]+aSha[1]+atou[1]+aWeapon[2]
                            charge = basic_char + aBei[4]+aHua[4]+aMao[4]+aSha[4]+atou[4]+aWeapon[5]+20
                            buff = charge*0.25 + (charge-100)*0.4 + aWeapon[4]+aBei[5]+27

                            damage = str*(1+buff/100)*(1+cri*criD/10000)*7.21*1.52*0.5*0.9
                            criDamage = str*(1+buff/100)*(1+criD/100)*7.21*1.52*0.5*0.9
                            # print("攻击力：", str)
                            # print("双爆：", cri, criD)
                            # print("recharge", charge)
                            # print("输出期望：", damage)
                            # print("武器", n+1)
                            # print("-----------------------------------------------")

                            if damage > best_damage:
                                best_damage = damage
                                best_equip = [i,j,k,l,m,n]
                                mianban = [str,cri, criD, charge]
                            if criDamage > big_damage and cri >= 50:
                                big_damage = criDamage
                                big_damage_mianban = [str,cri, criD, charge]
                                big_damage_equip = [i,j,k,l,m,n]

    return best_equip, best_damage, mianban,big_damage,big_damage_mianban,big_damage_equip
```

```python
def zhuanwuDamage():
    best_damage = 0
    best_equip = [0, 0, 0, 0, 0, 0]
    mianban = []
    big_damage = 0
    big_damage_mianban = []
    big_damage_equip = []
    for i, aHua in enumerate(hua):
        for j, aMao in enumerate(mao):
            for k, aSha in enumerate(sha):
                for l, atou in enumerate(tou):
                    for m, aBei in enumerate(bei):
                        charge = basic_char + aBei[4] + aHua[4] + aMao[4] + aSha[4] + atou[4] + 85 + 20
                        str = (basic_str + 608) * (
                                    (aBei[2] + aHua[2] + aMao[2] + aSha[2] + atou[2] + (charge-100)*0.28) / 100 + 1) + (
                                          aBei[3] + aHua[3] + aMao[3] + aSha[3] + atou[3])
                        cri = basic_cri + aBei[0] + aHua[0] + aMao[0] + aSha[0] + atou[0]
                        criD = basic_criD + aBei[1] + aHua[1] + aMao[1] + aSha[1] + atou[1]
                        if charge >= 300:
                            buff1 = 75
                        else:
                            buff1 = charge * 0.25

                        buff = buff1 + (charge - 100) * 0.4 + aBei[5] + 27
                        damage = str * (1 + buff / 100) * (1 + cri * criD / 10000) * 7.21 * 1.52 * 0.5 * 0.9
                        criDamage = str * (1 + buff / 100) * (1 + criD / 100) * 7.21 * 1.52 * 0.5 * 0.9
                        # print("攻击力：", str)
                        # print("双爆：", cri, criD)
                        # print("recharge", charge)
                        # print("输出期望：", damage)
                        # print("武器", n+1)
                        # print("-----------------------------------------------")

                        if damage > best_damage:
                            best_damage = damage
                            best_equip = [i , j , k , l , m ]
                            mianban = [str, cri, criD, charge]
                        if criDamage > big_damage and cri >= 50:
                            big_damage = criDamage
                            big_damage_mianban = [str, cri, criD, charge]
                            big_damage_equip = [i , j , k , l , m ]

    return best_equip, best_damage, mianban, big_damage, big_damage_mianban, big_damage_equip
```
