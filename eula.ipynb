```python
import numpy as np


basic_str = 342
basic_cri = 5
basic_criD = 88.4

set_1_buff = 50
set_2_buff_str = 18
set_2_buff_phy = 50


Skill = 7.256+1.482*13

emptyResult = [0,[],[]]
    
resultLen = 5
```

```python
# str, cri,crid,strbuff,defbuff,damageBUff,name

weapon = [
          [741,0,0,36,0,20.7,"1精专武"],
          [608,0,0,49.6+70+35,0,0,"4精狼末"],
          [608,0,0,49.6+40+20,0,0,"1精狼末"],
          [608,0,0,49.6+35,0,0,"4精狼末（无特效）"],
          [608,0,0,49.6+20,0,0,"1精狼末（无特效）"]
          ]

num_citiao = 28
```

```python
def get_damages():
#     套装攻物爆
    
    damage_result = [] 
    big_damage_result = []
    Set = ["2+2", "4苍白"]

    for n, aWeapon in enumerate(weapon):
        best_damage = 0
        best_equip = [0, 0, 0, 0, 0, 0]
        mianban = []
        big_damage = 0
        big_damage_mianban = []
        big_damage_equip = []
        weaponResult = []
        weaponBigRe = []
        for cri_num in range(num_citiao):
            for criD_num in range(num_citiao - cri_num):
                for aset in range(2):
                    str_num = num_citiao - cri_num - criD_num

                    cri = basic_cri+cri_num*3.3+aWeapon[1]+31.1
                    if cri > 100:
                        cri = 100
                    criD = basic_criD + criD_num*7 +aWeapon[2]

                    if aset == 0 :
                        buff = 58.7+ set_1_buff + 30+aWeapon[5]
                        Str = (basic_str+aWeapon[0])*((str_num*5.5+aWeapon[3])/100+1)+311+100
                    else:
                        buff = 58.7+ set_2_buff_phy + 30+aWeapon[5]
                        Str = (basic_str+aWeapon[0])*((str_num*5.5+aWeapon[3]+set_2_buff_str)/100+1)+311+100
                        
                    if "专武" in aWeapon[6]:                    
                        damage = (Str*Skill)*(1+buff/100)*(1+cri*criD/10000)*0.5*1.3
                        criDamage = (Str*Skill)*(1+buff/100)*(1+criD/100) *0.5*1.3
                    else:
                        damage = (Str*Skill)*(1+buff/100)*(1+cri*criD/10000)*0.5*1.3
                        criDamage = (Str*Skill)*(1+buff/100)*(1+criD/100) *0.5*1.3

                    if damage > best_damage and cri >60 and criD_num + cri_num  <= 20:
                        best_damage = damage
                        best_equip = [cri_num , criD_num , str_num  ,n ]
                        mianban = [Str, cri, criD,Set[aset]]
                    if criDamage > big_damage and cri >= 60 and criD_num + cri_num <= 20:
                        big_damage = criDamage
                        big_damage_mianban = [Str, cri, criD, Set[aset]]
                        big_damage_equip = [cri_num , criD_num , str_num  ,n  ]
        
        weaponResult.append(best_damage)
        weaponResult.append(best_equip)
        weaponResult.append(mianban)
        
        weaponBigRe.append(big_damage)
        weaponBigRe.append(big_damage_equip)
        weaponBigRe.append(big_damage_mianban)
        
        damage_result.append(weaponResult)
        big_damage_result.append(weaponBigRe)
        

    return damage_result, big_damage_result
```

```python
def Datasort(weapon_result):
    for i in range(len(weapon_result)):
        for j in range(i+1,len(weapon_result)):
            if weapon_result[j][0] > weapon_result[i][0]:
                aRe = []
                aRe = weapon_result[i]
                weapon_result[i] = weapon_result[j]
                weapon_result[j] = aRe
                
    return weapon_result
```

```python
damage_result, big_damage_result = get_damages()
damage_result = Datasort(damage_result)
big_damage_result = Datasort(big_damage_result)
best_damage = 0
best_cridamage = 0

for aResult in damage_result:
    if "专武" in weapon[aResult[1][3]][6]:
            best_damage = aResult[0]
            
for i,aResult in enumerate(damage_result):
    if i < 6:
        print("第"+str(i+1)+"输出面板：", aResult[2])

        print("暴击词条：", aResult[1][0])
        print("爆伤词条：", aResult[1][1])
        print("攻击词条：", aResult[1][2])
        print("配装武器：", weapon[aResult[1][3]][6])

        print("第"+str(i+1)+"输出期望", aResult[0])
        print("相比专武提升", round((aResult[0]/best_damage-1)*100,2), "%")
        print("--------------------------------------------------------------------")

print("\n")
print("\n")
for aResult in big_damage_result:
    if "专武" in weapon[aResult[1][3]][6]:
            best_damage = aResult[0]
for j,aResult in enumerate(big_damage_result):           
    if j < 6:
        print("第"+str(j+1)+"大数字面板：", aResult[2])

        print("暴击词条：", aResult[1][0])
        print("爆伤词条：", aResult[1][1])
        print("攻击词条：", aResult[1][2])
        print("配装武器：", weapon[aResult[1][3]][6])

        print("第"+str(j+1)+"输出大数字", aResult[0])
        if "专武" in weapon[aResult[1][3]][6]:
            best_damage = aResult[0]
        if i > 0:
            print("相比专武提升", round((aResult[0]/best_damage-1)*100,2), "%")
        print("--------------------------------------------------------------------")
```
