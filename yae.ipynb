```python
import numpy as np
import math

basic_str = 340
basic_cri = 24.2
basic_criD = 50


QSkill = 4.86+6.009*3

emptyResult = [0,[],[]]
    
resultLen = 5
```

```python
#weapon    0str,1cri,2crid,3str%,4Ebuff,5Qbuff,6mastrey,name
weapon = [[510,27.6,0,0,40*0.6,40,0,"满精匣里"],  
          [510,27.6,0,0,25*0.6,25,0,"2精匣里"],  
          [608,0,66.2,0,48,12,0,"1精专武"],
          [565,0,0,0,16,16,110,"满精海图"],
          [674,0,0,33.1,12,12,0,"天空书"],
          [565,0,0,27.6,0,0,0,"新法器"],
          [608,33.1,0,0,0,0,0,"四风"]
          ]

#set       0str%,1Qbuff,2Ebuff,name
Yiwu_set = [[36,0,0,"角斗追忆2+2"],
            [0,0,0,"绝缘4"],
            [18,20,0,"追忆宗室2+2"],
            [18,15,15,"追忆如雷2+2"],
             [0,35,15,"如雷宗室2+2"],
#             [0,35,35,"平雷4"]
           ]

num_citiao = 25
E_count = 20
E_Skills = [[2.133,"2命"],
            [1.706,"1命"]]

shalou = [[46.6,0,"攻击沙"],
          [0,51.6,"充能沙"]]
```

```python
def get_damages():
#     套装攻雷爆
    
    damage_result = [] 
    big_damage_result = []

    for aWeapon in weapon:
        best_damage = 0
        best_equip = [0, 0, 0, 0, 0, 0]
        mianban = []
        big_damage = 0
        big_damage_mianban = []
        big_damage_equip = []
        weaponResult = []
        weaponBigRe = []
        for aSet in Yiwu_set:
            
            for cri_num in range(num_citiao):
                for criD_num in range(num_citiao - cri_num):
                    for str_num in range(2,num_citiao - cri_num - criD_num):
                        for mastery_num in range(num_citiao - cri_num - criD_num - str_num):
                        
                            for ESkill in E_Skills:
                                for aSha in shalou:
                                    mastery = aWeapon[6] + 19.8*3
                                    cri = basic_cri+cri_num*3.3+aWeapon[1]
                                    if cri > 100:
                                        cri = 100
                                    criD = basic_criD + criD_num*6.6 +aWeapon[2]+62.2

                                    Str = (basic_str+aWeapon[0])*((str_num*4.95+aWeapon[3]+aSet[0]+aSha[0])/100+1)+311
                                    
                                    if aWeapon[7] == "新法器":
                                        recharge = 48+ aSha[1]+100
                                    else:
                                        recharge = 100+ aSha[1]

                                    Qbuff = 46.6 + aWeapon[5]+ aSet[1]
                                    if aSet == "绝缘":
                                        recharge += 20
                                        Qbuff += recharge*0.25
                                    Ebuff = 46.6 + aWeapon[4]+ aSet[2]+mastery*0.15

                                    Qdamage = (Str*QSkill)*(1+Qbuff/100)*(1+cri*criD/10000)*0.5*0.9
                                    QcriDamage = (Str*QSkill)*(1+Qbuff/100)*(1+criD/100) *0.5*0.9

                                    Edamage = (Str*ESkill[0])*(1+Ebuff/100)*(1+cri*criD/10000)*0.5*0.9
                                    EcriDamage = (Str*ESkill[0])*(1+Ebuff/100)*(1+criD/100) *0.5*0.9

                                    damage = Qdamage + E_count*Edamage
                                    criDamage = QcriDamage + E_count*EcriDamage

                                    if not (ESkill[1] == "2命" and "专武" in aWeapon[7]):
                                        if damage > best_damage  :
                                            best_damage = damage
                                            best_equip = [cri_num , criD_num , str_num  ,aWeapon ,mastery_num,aSet,ESkill,aSha]
                                            mianban = [Str, cri, criD, mastery]
                                        if criDamage > big_damage and cri >= 60 and criD_num + cri_num <= 18 and str_num < 10:
                                            big_damage = criDamage
                                            big_damage_mianban = [Str, cri, criD, mastery]
                                            big_damage_equip = [cri_num , criD_num , str_num ,aWeapon ,mastery_num,aSet,ESkill,aSha ]

        weaponResult.append(best_damage)
        weaponResult.append(best_equip)
        weaponResult.append(mianban)

        weaponBigRe.append(big_damage)
        weaponBigRe.append(big_damage_equip)
        weaponBigRe.append(big_damage_mianban)

        damage_result.append(weaponResult)
        big_damage_result.append(weaponBigRe)
        

    return damage_result, big_damage_result
```

```python
def Datasort(weapon_result):
    for i in range(len(weapon_result)):
        for j in range(i+1,len(weapon_result)):
            if weapon_result[j][0] > weapon_result[i][0]:
                aRe = []
                aRe = weapon_result[i]
                weapon_result[i] = weapon_result[j]
                weapon_result[j] = aRe
                
    return weapon_result
```

```python
damage_result, big_damage_result = get_damages()
damage_result = Datasort(damage_result)
big_damage_result = Datasort(big_damage_result)

for i,aResult in enumerate(damage_result):
    if i < 9:
        print("第"+str(i+1)+"输出面板：\n攻击力", aResult[2][0],"暴击率", aResult[2][1],"暴击伤害", aResult[2][2])
        print("暴击词条：", aResult[1][0],"爆伤词条：", aResult[1][1])
        print("攻击词条：", aResult[1][2])
        if aResult[1][4] > 0:
            print("精通词条：", aResult[1][4])
        print("配装武器：", aResult[1][3][7])
        print("配装套装：", aResult[1][5][3], aResult[1][7][2])
        print("选择命座：", aResult[1][6][1])
        
        print("第"+str(i+1)+"输出期望", aResult[0])
        
        if i == 0:
            best_damage = aResult[0]
        if i > 0:
            print("与第1输出期望差距", round((best_damage/aResult[0]-1)*100,2), "%")
        print("--------------------------------------------------------------------")

print("\n")
print("\n")
```
