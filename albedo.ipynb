```python
import numpy as np


basic_str = 251
basic_cri = 5
basic_criD = 50
basic_def = 876
basic_geo = 28.8

set_def_buff = 30+24
set_geo_buff = 24

# QSkill = 7.21+52*0.07+21.68
Skill = 2.2712

emptyResult = [0,[],[]]
    
resultLen = 5
```

```python
weapon = [[454,0,0,0,69,0,"专武"],
          [401,28,47,0,0,0,"黎明神剑"],
          [510,12,0,0,0,32,"腐殖之剑"]
          ]

num_citiao = 25
```

```python
def get_damages(gorou):
#     套装防岩爆
    
    damage_result = [] 
    big_damage_result = []
    Skill = 2.2712
    if not gorou:
        gorou_def_buff2 = 0
        gorou_5_criD = 0
    else:
        gorou_def_buff2 = 25
        gorou_5_criD = 40

    for n, aWeapon in enumerate(weapon):
        best_damage = 0
        best_equip = [0, 0, 0, 0, 0, 0]
        mianban = []
        big_damage = 0
        big_damage_mianban = []
        big_damage_equip = []
        weaponResult = []
        weaponBigRe = []
        for cri_num in range(num_citiao):
            for criD_num in range(num_citiao - cri_num):
                for str_num in range(num_citiao - cri_num - criD_num):
                    defence_num = num_citiao - cri_num - criD_num - str_num
   
                    Def = basic_def*(1+(defence_num*6 +aWeapon[4]+59+set_def_buff+gorou_def_buff2)/100)
                    cri = basic_cri+cri_num*3.3+aWeapon[1]
                    if cri > 100:
                        cri = 100
                    criD = basic_criD + criD_num*7 +aWeapon[2]+62.2

                    Str = (basic_str+aWeapon[0])*((str_num*5.5+aWeapon[3])/100+1)+311
                    buff = 46.6 +15 + aWeapon[5]+ set_geo_buff+ basic_geo
                
                    
                    if aWeapon[6] == "专武":
                        Skill = 2.2712 + 0.8
                    else:
                        Skill = 2.2712
                    
                    damage = Def*Skill*(1+buff/100)*(1+cri*criD/10000)*0.5*1.05
                    criDamage = Def*Skill*(1+buff/100)*(1+criD/100) *0.5*1.05


                    if damage > best_damage and cri >50 and criD_num + cri_num <= 20 :
                        best_damage = damage
                        best_equip = [cri_num , criD_num , str_num , defence_num ,n ]
                        mianban = [Str, cri, criD, Def]
                    if criDamage > big_damage and cri >= 50 and criD_num + cri_num <= 20:
                        big_damage = criDamage
                        big_damage_mianban = [Str, cri, criD, Def]
                        big_damage_equip = [cri_num , criD_num , str_num , defence_num ,n  ]
        
        weaponResult.append(best_damage)
        weaponResult.append(best_equip)
        weaponResult.append(mianban)
        
        weaponBigRe.append(big_damage)
        weaponBigRe.append(big_damage_equip)
        weaponBigRe.append(big_damage_mianban)
        
        damage_result.append(weaponResult)
        big_damage_result.append(weaponBigRe)
        

    return damage_result, big_damage_result
```

```python
def Datasort(weapon_result):
    for i in range(len(weapon_result)):
        for j in range(i+1,len(weapon_result)):
            if weapon_result[j][0] > weapon_result[i][0]:
                aRe = []
                aRe = weapon_result[i]
                weapon_result[i] = weapon_result[j]
                weapon_result[j] = aRe
                
    return weapon_result
```

```python
damage_result, big_damage_result = get_damages(True)
damage_result = Datasort(damage_result)
big_damage_result = Datasort(big_damage_result)

for i,aResult in enumerate(damage_result):
    print("第"+str(i+1)+"输出面板：", aResult[2])

    print("暴击词条：", aResult[1][0])
    print("爆伤词条：", aResult[1][1])
    print("攻击词条：", aResult[1][2])
    print("防御词条：", aResult[1][3])
    print("配装武器：", weapon[aResult[1][4]][6])

    print("第"+str(i+1)+"输出期望", aResult[0])
    print("--------------------------------------------------------------------")

print("\n")
print("\n")
for j,aResult in enumerate(big_damage_result):
    print("第"+str(j+1)+"大数字面板：", aResult[2])

    print("暴击词条：", aResult[1][0])
    print("爆伤词条：", aResult[1][1])
    print("攻击词条：", aResult[1][2])
    print("防御词条：", aResult[1][3])
    print("配装武器：", weapon[aResult[1][4]])

    print("第"+str(j+1)+"输出大数字", aResult[0])
    print("--------------------------------------------------------------------")
```
